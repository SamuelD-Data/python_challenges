# Classes and objects

```python
class favorite_animals(object): 
    def __init__(self, mammal, reptile, bird, fish): 
        self.mammal = mammal 
        self.reptile = reptile
        self.bird = bird
        self.fish = fish
 
    def print_fave_animals(self):
        print("My fave mammal is a", self.mammal)
        print("My fave reptile is a", self.reptile)
        print("My fave bird is a", self.bird)
        print("My fave fish is a", self.fish)
```

```python
first = favorite_animals('Koala', 'Snake', 'Eagle', 'Shark') 

first.reptile 
```

```python
first.print_fave_animals()
```

```python
getattr(first, "fish")
```

```python
setattr(first, "fish", "Whale") 

getattr(first, 'fish')
```

# Code-along for scientific calculator

```python
import math

while True:

    print("\nChoose the math operation you'd like to perform.\n\n0 - Addition\n1 - Subtraction\n2 - Multiplication\n3 - Division\n4 - Modulo\n5 - Raising to a power\n6 - Square root\n7 - Logarithm\n8 - Sine\n9 - Cosine\n10 - Tangent\n")

    oper = input("\nYour option from the menu: ") 

    if oper == "0":
        val1 = float(input("\Initial value: ")) 
        val2 = float(input("\Secondary value: ")) 

        print("\nThe result is: " + str(val1 + val2) + "\n")

        back = input('\nWould you like to go back to the main menu? (y/n) ')

        if back == 'y':
            continue
        else:
            break

    elif oper == "1":
        val1 = float(input("\Initial value: "))
        val2 = float(input("\nSecondary value: "))

        print("\nThe result is: " + str(val1 - val2) + "\n")

        back = input('\nWould you like to go back to the main menu? (y/n) ')

        if back == 'y':
            continue
        else:
            break

    elif oper == "2":
        val1 = float(input("\Initial value: "))
        val2 = float(input("\nSecondary value: "))

        print("\nThe result is: " + str(val1 * val2) + "\n")

        back = input('\nWould you like to go back to the main menu? (y/n) ')

        if back == 'y':
            continue
        else:
            break

    elif oper == "3":
        val1 = float(input("\Initial value: "))
        val2 = float(input("\nSecondary value: "))

        print("\nThe result is: " + str(val1 / val2) + "\n")

        back = input('\nWould you back to the main menu? (y/n) ')

        if back == 'y':
            continue
        else:
            break

    elif oper == "4":
        val1 = float(input("\nInitial value: "))
        val2 = float(input("\nSecondary value: "))

        print("\nThe result is: " + str(val1 % val2) + "\n")

        back = input('\nWould you like to go back to the main menu? (y/n) ')

        if back == 'y':
            continue
        else:
            break

    elif oper == "5":
        val1 = float(input("\nPlease enter the base: "))
        val2 = float(input("\nPlease enter the power: "))

        print("\nThe result is: " + str(math.pow(val1, val2)) + "\n")

        back = input('\nWould you like to go back to the main menu? (y/n) ')

        if back == 'y':
            continue
        else:
            break

    elif oper == "6":
        val1 = float(input("\nEnter a value to calculate the square root from: "))

        print("\nThe result is: " + str(math.sqrt(val1)) + "\n")

        back = input('\nWould you like to go back to the main menu? (y/n) ')

        if back == 'y':
            continue
        else:
            break
    
    elif oper == "7":
        val1 = float(input("\nEnter the value for calculating the logarithm to base 2: "))

        print("\nThe result is: " + str(math.log(val1, 2)) + "\n")

        back = input('\nWould you like to go back to the main menu? (y/n) ')

        if back == 'y':
            continue
        else:
            break
    
    elif oper == "8":
        val1 = float(input("\nEnter the value for calculating the sine: "))

        print("\nThe result is: " + str(math.sin(math.radians(val1))) + "\n")

        back = input('\nWould you like to go back to the main menu? (y/n) ')

        if back == 'y':
            continue
        else:
            break

    elif oper == "9":
        val1 = float(input("\nEnter the value to calculate the cosign of: "))

        print("\nThe result is: " + str(math.cos(math.radians(val1))) + "\n")

        back = input('\nWould you like to go back to the main menu? (y/n) ')

        if back == 'y':
            continue
        else:
            break

    elif oper == "10":
        val1 = float(input("\nEnter the value to calculate the tangent of: "))

        print("\nThe result is: " + str(math.tan(math.radians(val1))) + "\n")

        back = input('\nWould you like to go back to the main menu? (y/n) ')

        if back == 'y':
            continue
        else:
            break
    
    else:
        print("\nThat is an invalid option\n")
        continue
```
