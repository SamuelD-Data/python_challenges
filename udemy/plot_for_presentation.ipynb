```python
df = pd.DataFrame({'neutral': [.59, .66], 'positive': [.58, .64], 'negative':[.57,.61]})

plt.figure(figsize = (20,10)) 
lines = df.plot.line(color = {'green':'neutral','blue':'positive','red':'negative'})
lines.set_xticks([0,1])
lines.set_xticklabels(['Before Program', 'After Program'], size = 20)
lines.scatter([0,1],[.59, .66], color = 'g')
lines.scatter([0,1],[.58, .64], color = 'b')
lines.scatter([0,1],[.57, .61], color = 'r')
plt.suptitle("Student Performance", y = .93, fontsize = 25)
plt.ylabel('Percent of Correct Answers', size = 20)
plt.rc('ytick', labelsize = 20)
plt.show()
```
