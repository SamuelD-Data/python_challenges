__1) Write a function named capital_indexes. The function takes a single parameter, which is a string. Your function should return a list of all the indexes in the string that have capital letters. For example, calling capital_indexes("HeLlO") should return the list [0, 2, 4].__

```python
word = "HeLlO"

word.find('e')
```

```python
capitals = []

for x in word:
    if x == x.upper():
        capitals.append(word.find(x))
        
capitals    
```

```python
def capital_indexes(word):
    capitals = []
    for x in word:
        if x == x.upper():
            capitals.append(word.find(x))
    return capitals

capital_indexes(word)
```

```python
enum_list = []

for place, letter in enumerate(word):
    enum_list.append([place, letter])

enum_list
```

```python
capital_places = []

for place, letter in enumerate(word):
    if letter == letter.upper():
        capital_places.append(place)

capital_places
```
