__1) Write a function named capital_indexes. The function takes a single parameter, which is a string. Your function should return a list of all the indexes in the string that have capital letters. For example, calling capital_indexes("HeLlO") should return the list [0, 2, 4].__

```python
word = "HeLlO"

word.find('e')
```

```python
capitals = []

for x in word:
    if x == x.upper():
        capitals.append(word.find(x))
        
capitals    
```

```python
def capital_indexes(word):
    capitals = []
    for x in word:
        if x == x.upper():
            capitals.append(word.find(x))
    return capitals

capital_indexes(word)
```

```python
def capital_indexes2(word):
    capital_places = []
    for place, letter in enumerate(word):
        if letter == letter.upper():
            capital_places.append(place)
    return capital_places

capital_indexes2('TeST')
```

__2) Write a function named mid that takes a string as its parameter. Your function should extract and return the middle letter. 
If there is no middle letter, your function should return the empty string.
For example, mid("abc") should return "b" and mid("aaaa") should return "".__

```python
import math

def mid(word):
    if len(word) % 2 == 0:
        return ""
    else:
        mid_pos = math.floor(len(word) / 2)
        return word[mid_pos]
    
        
```

```python
mid('python')
```

```python
mid('kites')
```
