# Dictionary Problems

1. Write a Python program to sort (ascending and descending) a dictionary by value. 

```python
d = {'a': 4, 'b': 80, 'c': 10, 'd': 2, 'e': 66, 'f': 1}

l = []

for x in d.items():
    l.append(x)
    
l = sorted(l, key=lambda x: x[1])

dict_list = dict(l)

dict_list
```

2. Write a Python program to add a key to a dictionary. 

Sample Dictionary : {0: 10, 1: 20}
Expected Result : {0: 10, 1: 20, 2: 30}

```python
d = {0: 10, 1: 20}

def addkey(d, k, v):
    d[k] = v
    return d

addkey(d, 2, 30)
```

3. Write a Python program to concatenate following dictionaries to create a new one. 

Sample Dictionary : 
dic1={1:10, 2:20} 
dic2={3:30, 4:40} 
dic3={5:50,6:60}
Expected Result : {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}

```python
dic1 = {1:10, 2:20} 
dic2 = {3:30, 4:40} 
dic3 = {5:50,6:60}

def combine_d3(a, b, c):
    a.update(b)
    a.update(c)
    return a

combine_d3(dic1, dic2, dic3)
```
