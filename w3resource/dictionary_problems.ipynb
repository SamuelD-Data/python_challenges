1. Write a Python program to sort (ascending and descending) a dictionary by value. 

```python
d = {'a': 4, 'b': 80, 'c': 10, 'd': 2, 'e': 66, 'f': 1}

l = []

for x in d.items():
    l.append(x)
    
l = sorted(l, key=lambda x: x[1])

dict_list = dict(l)

dict_list
```

4. Write a Python program to check whether a given key already exists in a dictionary. 

```python
d = {'a': 4, 'b': 80, 'c': 10, 'd': 2, 'e': 66, 'f': 1}

def check_key(d, k):
    status = True
    for x in d:
        if x == k:
            status = False
    if status == False:
        print('key already exists')
    else:
        print('key does not exist')

check_key(d, 'b')
```

6. Write a Python program to generate and print a dictionary that contains a number (between 1 and n) in the form (x, x*x). 
Sample Dictionary ( n = 5) : 
Expected Output : {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

```python
def make_dict(n):
    d = {}
    for x in range(1, n+1):
        d[x] = x * x
    return d
        
make_dict(5)
```

10. Write a Python program to sum all the items in a dictionary. 

```python
d = {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}

def sum_keys_and_vals(a):
    keysum = 0
    vsum = 0
    osum = 0
    for x in a.items():
        keysum += x[0]
        vsum += x[1]
        osum += (x[0] + x[1])
    print('sum of keys:', keysum, 'sum of values:', vsum, 'sum of keys and values:', osum)
        
sum_keys_and_vals(d)
        
```

11. Write a Python program to multiply all the items in a dictionary. 

```python
d = {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}

def prod_keys_and_vals(a):
    keyprod = 1
    vprod = 1
    oprod = 1
    for x in a.items():
        keyprod *= x[0]
        vprod *= x[1]
        oprod *= (x[0] + x[1])
    print('product of keys:', keyprod, 'product of values:', vprod, 'product of keys and values:', oprod)
        
prod_keys_and_vals(d)
```

13. Write a Python program to map two lists into a dictionary. 

```python
a = [1,2,3,4]
b = ['a', 'b', 'c', 'd']

def map_dict(l1, l2):
    d = {}
    for x in l1:
        for y in l2:
            d[x] = y
    return d

map_dict(a, b)
```

19. Write a Python program to combine two dictionary adding values for common keys. 
d1 = {'a': 100, 'b': 200, 'c':300}
d2 = {'a': 300, 'b': 200, 'd':400}
Sample output: Counter({'a': 400, 'b': 400, 'd': 400, 'c': 300})

```python
d1 = {'a': 100, 'b': 200, 'c':300}
d2 = {'a': 300, 'b': 200, 'd':400}

def add_dicts(a, b):
    for x in a:
        if x in b:
            a[x] += b[x]
    for y in b:
        if y not in a:
            a[y] = b[y]
    return a
            
add_dicts(d1, d2)
```

20. Write a Python program to print all unique values from a list of dictionaroes. 
Sample Data : [{"V":"S001"}, {"V": "S002"}, {"VI": "S001"}, {"VI": "S005"}, {"VII":"S005"}, {"V":"S009"},{"VIII":"S007"}]
Expected Output : Unique Values: {'S005', 'S002', 'S007', 'S001', 'S009'}

```python
d = [{"V":"S001"}, {"V": "S002"}, {"VI": "S001"}, {"VI": "S005"}, {"VII":"S005"}, {"V":"S009"},{"VIII":"S007"}]

def unique_vals(a):
    l = []
    l2 = []
    for x in a:
        l.append(list(x.values())) 
    for y in l:
        l2.append(y[0])
    return (set(l2))

unique_vals(d)
```

21. Write a Python program to create and display all combinations of letters, selecting each letter from a different key in a dictionary. 
Sample data : {'1':['a','b'], '2':['c','d']}
Expected Output: 
ac
ad
bc
bd

```python
d = {'1':['a','b'], '2':['c','d']}

def d_combo(d):
    
    l = []
    l2 = []

    for x in d.values():
        l.append(x)

    for y in enumerate(l):
        for z in enumerate(l):
            if y[0] != z[0]:
                l2.append((y[1],z[1]))
    
    return l2
        
d_combo(d)
```

22. Write a Python program to find the highest n values in a dictionary. 

```python
d = {'a': 4, 'b': 80, 'c': 10, 'd': 2, 'e': 66, 'f': 1, 'a': 4}

def highest_n(d, n):

    l = []
    
    for x in d.values():
        l.append(x)

    l.sort(reverse = True)
    l = l[0:n]

    return l

highest_n(d, 3)
```
