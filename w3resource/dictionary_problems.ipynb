10. Write a Python program to sum all the items in a dictionary. 

```python
d = {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}

def sum_keys_and_vals(a):
    keysum = 0
    vsum = 0
    osum = 0
    for x in a.items():
        keysum += x[0]
        vsum += x[1]
        osum += (x[0] + x[1])
    print('sum of keys:', keysum, 'sum of values:', vsum, 'sum of keys and values:', osum)
        
sum_keys_and_vals(d)
        
```

11. Write a Python program to multiply all the items in a dictionary. 

```python
d = {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}

def prod_keys_and_vals(a):
    keyprod = 1
    vprod = 1
    oprod = 1
    for x in a.items():
        keyprod *= x[0]
        vprod *= x[1]
        oprod *= (x[0] + x[1])
    print('product of keys:', keyprod, 'product of values:', vprod, 'product of keys and values:', oprod)
        
prod_keys_and_vals(d)
```

19. Write a Python program to combine two dictionary adding values for common keys. 
d1 = {'a': 100, 'b': 200, 'c':300}
d2 = {'a': 300, 'b': 200, 'd':400}
Sample output: Counter({'a': 400, 'b': 400, 'd': 400, 'c': 300})

```python
d1 = {'a': 100, 'b': 200, 'c':300}
d2 = {'a': 300, 'b': 200, 'd':400}

def add_dicts(a, b):
    for x in a:
        if x in b:
            a[x] += b[x]
    for y in b:
        if y not in a:
            a[y] = b[y]
    return a
            
add_dicts(d1, d2)
```

21. Write a Python program to create and display all combinations of letters, selecting each letter from a different key in a dictionary. 
Sample data : {'1':['a','b'], '2':['c','d']}
Expected Output: 
ac
ad
bc
bd

```python
d = {'1':['a','b'], '2':['c','d']}

def d_combo(d):
    
    l = []
    l2 = []

    for x in d.values():
        l.append(x)

    for y in enumerate(l):
        for z in enumerate(l):
            if y[0] != z[0]:
                l2.append((y[1],z[1]))
    
    return l2
        
d_combo(d)
```

22. Write a Python program to find the highest n values in a dictionary. 

```python
d = {'a': 4, 'b': 80, 'c': 10, 'd': 2, 'e': 66, 'f': 1, 'a': 4}

def highest_n(d, n):

    l = []
    
    for x in d.values():
        l.append(x)

    l.sort(reverse = True)
    l = l[0:n]

    return l

highest_n(d, 3)
```

43. Write a Python program to convert more than one list to nested dictionary. 
Original strings:
['S001', 'S002', 'S003', 'S004']
['Adina Park', 'Leyton Marsh', 'Duncan Boyle', 'Saim Richards']
[85, 98, 89, 92]
Nested dictionary:
[{'S001': {'Adina Park': 85}}, {'S002': {'Leyton Marsh': 98}}, {'S003': {'Duncan Boyle': 89}}, {'S004': {'Saim Richards': 92}}]

```python
l1 = ['S001', 'S002', 'S003', 'S004'] 
l2 = ['Adina Park', 'Leyton Marsh', 'Duncan Boyle', 'Saim Richards'] 
l3 = [85, 98, 89, 92] 
d = {}

ziplist = dict(zip(l2, l3))

for x in enumerate(l1):
    for y in enumerate(ziplist):
        if x[0] == y[0]:
            d[x[1]] = ziplist[y[1]]
        
d
```

45. Write a Python program to check all values are same in a dictionary. 
Original Dictionary:
{'Cierra Vega': 12, 'Alden Cantrell': 12, 'Kierra Gentry': 12, 'Pierre Cox': 12}
Check all are 12 in the dictionary.
True
Check all are 10 in the dictionary.
False

```python
d = {'Cierra Vega': 12, 'Alden Cantrell': 12, 'Kierra Gentry': 12, 'Pierre Cox': 12}

def match_val_check(d, n):
    vals = [x for x in d.values()]
    vals = list(set(vals))
    if len(vals) > 1:
        print('not all values match')
    elif vals[0] != n:
        print('values match but are not ' + str(n))
    else:
        print('values all match and are equal to ' + str(n))
        
match_val_check(d, 12)
```

51. A Python Dictionary contains List as value. Write a Python program to update the list values in the said dictionary. 
Original Dictionary:
{'Math': [88, 89, 90], 'Physics': [92, 94, 89], 'Chemistry': [90, 87, 93]}
Update the list values of the said dictionary:
{'Math': [89, 90, 91], 'Physics': [90, 92, 87], 'Chemistry': [90, 87, 93]}

```python
d = {'Math': [88, 89, 90], 'Physics': [92, 94, 89], 'Chemistry': [90, 87, 93]}

def update_dict(d, k, v):
    for x in enumerate(k):
        for y in enumerate(v):
            if x[0] == y[0]:
                d[x[1]] = y[1]
    return d

update_dict(d, ['Math', 'Physics'], [[89, 90, 91],[90, 92, 87]])

```
