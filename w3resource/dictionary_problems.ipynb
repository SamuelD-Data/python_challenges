# Dictionary Problems

1. Write a Python program to sort (ascending and descending) a dictionary by value. 

```python
d = {'a': 4, 'b': 80, 'c': 10, 'd': 2, 'e': 66, 'f': 1}

l = []

for x in d.items():
    l.append(x)
    
l = sorted(l, key=lambda x: x[1])

dict_list = dict(l)

dict_list
```

2. Write a Python program to add a key to a dictionary. 

Sample Dictionary : {0: 10, 1: 20}
Expected Result : {0: 10, 1: 20, 2: 30}

```python
d = {0: 10, 1: 20}

def addkey(d, k, v):
    d[k] = v
    return d

addkey(d, 2, 30)
```

3. Write a Python program to concatenate following dictionaries to create a new one. 

Sample Dictionary : 
dic1={1:10, 2:20} 
dic2={3:30, 4:40} 
dic3={5:50,6:60}
Expected Result : {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}

```python
dic1 = {1:10, 2:20} 
dic2 = {3:30, 4:40} 
dic3 = {5:50,6:60}

def combine_d3(a, b, c):
    a.update(b)
    a.update(c)
    return a

combine_d3(dic1, dic2, dic3)
```

4. Write a Python program to check whether a given key already exists in a dictionary. 

```python
d = {'a': 4, 'b': 80, 'c': 10, 'd': 2, 'e': 66, 'f': 1}

def check_key(d, k):
    status = True
    for x in d:
        if x == k:
            status = False
    if status == False:
        print('key already exists')
    else:
        print('key does not exist')

check_key(d, 'b')
```

5. Write a Python program to iterate over dictionaries using for loops. 

```python
d = {'a': 4, 'b': 80, 'c': 10, 'd': 2, 'e': 66, 'f': 1}
d2 = {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}

def iterate_dicts(a, b):
    for x in a:
        for y in b:
            print(x, y)
            
iterate_dicts(d, d2)
```

6. Write a Python program to generate and print a dictionary that contains a number (between 1 and n) in the form (x, x*x). 
Sample Dictionary ( n = 5) : 
Expected Output : {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

```python
def make_dict(n):
    d = {}
    for x in range(1, n+1):
        d[x] = x * x
    return d
        
make_dict(5)
```
