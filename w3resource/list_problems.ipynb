```python
one = ['cat', 'bat', 'moth']
two = ['rat', 'kangaroo', 'moth']

def listcheck(l1, l2):
    status = False
    
    for x in l1:
        if x in l2:
            status = True
            
    return status

listcheck(one, two)
```

31. Write a Python program to count the number of elements in a list within a specified range. 

```python
def rangecount(l,r):
    
    counter = 0
    
    for x in l:
        if x >= r[0] and x <= r[1]:
            counter += 1
    
    print(counter)
    
rangecount([1,3,4,5,6,4,2,4,67,52,46], [1,10])
        
```

35. Write a Python program to create a list by concatenating a given list which range goes from 1 to n. 
Sample list : ['p', 'q']
n =5
Sample Output : ['p1', 'q1', 'p2', 'q2', 'p3', 'q3', 'p4', 'q4', 'p5', 'q5']

```python
def itemcat(l, n):
    minilist = []
    for i in range(1, n+1):
        for x in l:
            minilist.append(x + str(i))
    print(minilist)
            
itemcat(['p','q'], 2)
```
