4. Write a Python program to get a string from a given string where all occurrences of its first char have been changed to '$', except the first char itself.  

```python
def afterfirst(s):
    rest = s[1:]
    rest = rest.replace(s[0], '$')
    return s[0] + rest
    
afterfirst('lollipop')
```

8. Write a Python function that takes a list of words and returns the longest word and the length of the longest one.  
Sample Output:
Longest word: Exercises
Length of the longest word: 9

```python
def wordl(l):
    maxlen = max([len(x) for x in l])
    return [y for y in l if len(y) == maxlen], maxlen

wordl(['bob', 'cat', 'sheep', 'cow'])
```

Sample function and result :

20. Write a Python function to reverses a string if it's length is a multiple of 4. 

```python
def reversefour(s):
    lencheck = len(s)
    if lencheck % 4 == 0:
        return s[::-1]
    else:
        return s
    
reversefour('catering')
```

21. Write a Python function to convert a given string to all uppercase if it contains at least 2 uppercase characters in the first 4 characters. 

```python
def uppertwofour(s):
    ff = s[:4]
    uppercounter = 0
    for x in ff:
        if x == x.upper():
            uppercounter += 1
    if uppercounter > 1:
        return s.upper()
    else:
        return s

uppertwofour('HYper')
```

Sample string: 'thequickbrownfoxjumpsoverthelazydog'
Expected output :
o 4
e 3
u 2
h 2
r 2
t 2

```python
s = 'thequickbrownfoxjumpsoverthelazydog'

wdict = {}

for x in s:
    if x not in wdict.keys():
        wdict[x] = 1
    else:
        wdict[x] += 1
    
for x in wdict:
    if wdict[x] > 1:
        print(x, wdict[x])
    
```
