16. Write a Python program to add the digits of a positive integer repeatedly until the result has a single digit. 
Input : 48
Output : 3
For example given number is 59, the result will be 5.
Step 1: 5 + 9 = 14
Step 1: 1 + 4 = 5



```python
def untilsingle(n):
    while len(str(n)) > 1:
        n = str(n)
        n = [c for c in n]
        n = [int(c) for c in n]
        n = sum(n)
    return n

untilsingle(99)
```

Input : 'anagram','nagaram'

```python
def check_ana(s1, s2):
    s1 = [c for c in s1]
    s2 = [c for c in s2]
    s1.sort(), s2.sort()
    if s1 == s2:
        print('both words are anagrams of each other')
    else:
        print('the words are not anagrams of eacho ther')
    
check_ana('tac', 'cat')
```

A Caesar cipher is a simple substitution cipher in which each letter of the plain text is substituted with a letter found by moving n places down the alphabet. You are to write a function that accepts two arguments, a plain-text message and a number of letters to shift in the cipher. The function will return an encrypted string with all letters transformed and all punctuation and whitespace remaining unchanged.

```python
import string 

alphabet = string.ascii_lowercase

aenum = dict([x for x in enumerate(alphabet)])

s = 'abd'
nl = []

for l in s:
    for c in aenum:
        if l == aenum[c]:
            nl.append(c)
            
nl2 = []

for n in nl:
    nl2.append(aenum[n])
    
es = ''
for x in nl2:
    es += x
    
es
```

```python
def cypher(s, n):
    
    alphabet = string.ascii_lowercase

    aenum = dict([x for x in enumerate(alphabet)])
    
    nl = []
    
    for l in s:
        for c in aenum:
            if l == aenum[c]:
                nl.append(c + n)
                
    nl2 = []
    
    for n in nl:
        nl2.append(aenum[n])
        
    es = ''
    
    for x in nl2:
        es += x
        
    return es
```

```python
cypher('hi', 2)
```

Make a 2-player rock, paper, scissors game

```python
def rock_paper_sci():
    
    p1 = input()
    p2 = input()
    
    if p1.lower() == 'rock' and p2.lower() == 'scissors':
        print('player 1 wins')
        
    elif p1.lower() == 'scissors' and p2.lower() == 'paper':
        print('player 1 wins')
        
    elif p1.lower() == 'paper' and p2.lower() == 'rock':
        print('player 1 wins')
        
    elif p1.lower() == p2.lower():
        print('Tie!')
        
    else:
        print('player 2 wins')
        
rock_paper_sci()
```

```python
def rock_paper_sci():
    
    import random
    
    p1 = input()
    comp = random.choice(['rock', 'paper', 'scissors'])
    
    if p1.lower() == 'rock' and comp.lower() == 'scissors':
        print('computer picked', comp)
        print('player 1 wins')
        
    elif p1.lower() == 'scissors' and comp.lower() == 'paper':
        print('player 1 wins')
        
    elif p1.lower() == 'paper' and comp.lower() == 'rock':
        print('player 1 wins')
        
    elif p1.lower() == comp.lower():
        print('Tie!')
        
    else:
        print('computer picked', comp)
        print('computer wins')
        
rock_paper_sci()
```

Generate a random number between 1 and 9 (including 1 and 9). Ask the user to guess the number, then tell them whether they guessed too low, too high, or exactly right. (Hint: remember to use the user input lessons from the very first exercise)

Extras:

Keep the game going until the user types “exit”
Keep track of how many guesses the user has taken, and when the game ends, print this out.

```python
import random 

def guess_game():
    comp_number = random.choice(range(1,9))
    counter = 0
    
    while 1 == 1:
        counter += 1
        guess = input('Enter a number between 1 and 9! ')
        if guess.lower() == 'exit':
            print('exiting game')
            break
        elif int(guess) > comp_number:
            print('Too high! Guess again')
            print('Number of guesses: ', counter)
        elif int(guess) < comp_number:
            print('Too low! Guess again')
            print('Number of guesses: ', counter)
        else:
            print('Youre right!')
            print('Number of guesses: ', counter)
            break

guess_game()
```
