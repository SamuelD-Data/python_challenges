16. Write a Python program to add the digits of a positive integer repeatedly until the result has a single digit. 
Input : 48
Output : 3
For example given number is 59, the result will be 5.
Step 1: 5 + 9 = 14
Step 1: 1 + 4 = 5



```python
def untilsingle(n):
    while len(str(n)) > 1:
        n = str(n)
        n = [c for c in n]
        n = [int(c) for c in n]
        n = sum(n)
    return n

untilsingle(99)
```

26. Write a Python program to check if a given string is an anagram of another given string.
Input : 'anagram','nagaram'
Output : True 

```python
def check_ana(s1, s2):
    s1 = [c for c in s1]
    s2 = [c for c in s2]
    s1.sort(), s2.sort()
    if s1 == s2:
        print('both words are anagrams of each other')
    else:
        print('the words are not anagrams of eacho ther')
    
check_ana('tac', 'cat')
```

4. Create a Python project that prints out every line of the song "99 bottles of soda on the wall." Note: Try to use a built in function instead of manually type all the lines.

```python
def soda_song():
    n = int(input('how many bottles of soda should we start with? '))
    while n >= 1:
        print(n, 'bottles of soda of the wall,', n, 'bottles of soda!')
        n -= 1
        if n == 0:
            print('No more soda!')

soda_song()
      
```
