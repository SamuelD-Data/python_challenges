16. Write a Python program to add the digits of a positive integer repeatedly until the result has a single digit. 
Input : 48
Output : 3
For example given number is 59, the result will be 5.
Step 1: 5 + 9 = 14
Step 1: 1 + 4 = 5



```python
def untilsingle(n):
    while len(str(n)) > 1:
        n = str(n)
        n = [c for c in n]
        n = [int(c) for c in n]
        n = sum(n)
    return n

untilsingle(99)
```

A Caesar cipher is a simple substitution cipher in which each letter of the plain text is substituted with a letter found by moving n places down the alphabet. You are to write a function that accepts two arguments, a plain-text message and a number of letters to shift in the cipher. The function will return an encrypted string with all letters transformed and all punctuation and whitespace remaining unchanged.

```python
import string 

alphabet = string.ascii_lowercase

aenum = dict([x for x in enumerate(alphabet)])

s = 'abd'
nl = []

for l in s:
    for c in aenum:
        if l == aenum[c]:
            nl.append(c)
            
nl2 = []

for n in nl:
    nl2.append(aenum[n])
    
es = ''
for x in nl2:
    es += x
    
es
```

```python
def cypher(s, n):
    
    alphabet = string.ascii_lowercase

    aenum = dict([x for x in enumerate(alphabet)])
    
    nl = []
    
    for l in s:
        for c in aenum:
            if l == aenum[c]:
                nl.append(c + n)
                
    nl2 = []
    
    for n in nl:
        nl2.append(aenum[n])
        
    es = ''
    
    for x in nl2:
        es += x
        
    return es
```

```python
cypher('hi', 2)
```

Extras:

Keep the game going until the user types “exit”
Keep track of how many guesses the user has taken, and when the game ends, print this out.

```python
import random 

def guess_game():
    comp_number = random.choice(range(1,9))
    counter = 0
    
    while 1 == 1:
        counter += 1
        guess = input('Enter a number between 1 and 9! ')
        if guess.lower() == 'exit':
            print('exiting game')
            break
        elif int(guess) > comp_number:
            print('Too high! Guess again')
            print('Number of guesses: ', counter)
        elif int(guess) < comp_number:
            print('Too low! Guess again')
            print('Number of guesses: ', counter)
        else:
            print('Youre right!')
            print('Number of guesses: ', counter)
            break

guess_game()
```

Write a program that accepts a sentence and calculate the number of letters and digits.
Suppose the following input is supplied to the program:
hello world! 123
Then, the output should be:
LETTERS 10
DIGITS 3

```python
import string

def letters_and_digits(s):
    digits = 0
    letters = 0
    for x in s:
        if x in string.ascii_lowercase:
            letters += 1
        elif x in '1234567890':
            digits += 1
    print('LETTERS:', letters)
    print('DIGITS:', digits)
    
letters_and_digits('hello world! 123')
```

Write a program that computes the net amount of a bank account based a transaction log from console input. 

```python
def bank(*args):
    total = 0
    for x in args:
        if x[0] == 'w':
            total -= int(x[2:])
        else:
            total += int(x[2:])
    print(total)
        
bank('w 200', 'd 300', 'w 50')
```

Please write assert statements to verify that every number in the list [2,4,6,8] is even, odd, or mixed.

```python
def check_signs(l):
    evens = []
    odds = []
    for x in l:
        if x % 2 == 0:
            evens.append(x)
        elif x % 2 == 1:
            odds.append(x)
    if len(evens) == len(l):
        print('all numbers are even')
    elif len(odds) == len(l):
        print('all numbers are odd')
    else:
        print('list contains mix of odds and evens')
        
check_signs([2,4,6,8])
```

Create a function that returns a string in pig latin.

```python
def pig_latin(s):
    sentence = []
    for x in s.split():
        fl = x[0]
        rest = x[1:]
        sentence.append(rest + fl + 'ay')
    return ' '.join(sentence)

pig_latin('hello world im someone')
```
