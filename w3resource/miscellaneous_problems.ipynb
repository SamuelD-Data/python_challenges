16. Write a Python program to add the digits of a positive integer repeatedly until the result has a single digit. 
Input : 48
Output : 3
For example given number is 59, the result will be 5.
Step 1: 5 + 9 = 14
Step 1: 1 + 4 = 5



```python
def untilsingle(n):
    while len(str(n)) > 1:
        n = str(n)
        n = [c for c in n]
        n = [int(c) for c in n]
        n = sum(n)
    return n

untilsingle(99)
```

26. Write a Python program to check if a given string is an anagram of another given string.
Input : 'anagram','nagaram'
Output : True 

```python
def check_ana(s1, s2):
    s1 = [c for c in s1]
    s2 = [c for c in s2]
    s1.sort(), s2.sort()
    if s1 == s2:
        print('both words are anagrams of each other')
    else:
        print('the words are not anagrams of eacho ther')
    
check_ana('tac', 'cat')
```

4. Create a Python project that prints out every line of the song "99 bottles of soda on the wall." Note: Try to use a built in function instead of manually type all the lines.

```python
def soda_song():
    n = int(input('how many bottles of soda should we start with? '))
    while n >= 1:
        print(n, 'bottles of soda of the wall,', n, 'bottles of soda!')
        n -= 1
        if n == 0:
            print('No more soda!')

soda_song()
      
```

A Caesar cipher is a simple substitution cipher in which each letter of the plain text is substituted with a letter found by moving n places down the alphabet. You are to write a function that accepts two arguments, a plain-text message and a number of letters to shift in the cipher. The function will return an encrypted string with all letters transformed and all punctuation and whitespace remaining unchanged.

```python
import string 

alphabet = string.ascii_lowercase

aenum = dict([x for x in enumerate(alphabet)])

s = 'abd'
nl = []

for l in s:
    for c in aenum:
        if l == aenum[c]:
            nl.append(c)
            
nl2 = []

for n in nl:
    nl2.append(aenum[n])
    
es = ''
for x in nl2:
    es += x
    
es
```

```python
def cypher(s, n):
    
    alphabet = string.ascii_lowercase

    aenum = dict([x for x in enumerate(alphabet)])
    
    nl = []
    
    for l in s:
        for c in aenum:
            if l == aenum[c]:
                nl.append(c + n)
                
    nl2 = []
    
    for n in nl:
        nl2.append(aenum[n])
        
    es = ''
    
    for x in nl2:
        es += x
        
    return es
```

```python
cypher('hi', 2)
```

Make a 2-player rock, paper, scissors game

```python
def rock_paper_sci():
    
    p1 = input()
    p2 = input()
    
    if p1.lower() == 'rock' and p2.lower() == 'scissors':
        print('player 1 wins')
        
    elif p1.lower() == 'scissors' and p2.lower() == 'paper':
        print('player 1 wins')
        
    elif p1.lower() == 'paper' and p2.lower() == 'rock':
        print('player 1 wins')
        
    elif p1.lower() == p2.lower():
        print('Tie!')
        
    else:
        print('player 2 wins')
        
rock_paper_sci()
```

Make a 1-player rock, paper, scissors game

```python
def rock_paper_sci():
    
    import random
    
    p1 = input()
    comp = random.choice(['rock', 'paper', 'scissors'])
    
    if p1.lower() == 'rock' and comp.lower() == 'scissors':
        print('computer picked', comp)
        print('player 1 wins')
        
    elif p1.lower() == 'scissors' and comp.lower() == 'paper':
        print('player 1 wins')
        
    elif p1.lower() == 'paper' and comp.lower() == 'rock':
        print('player 1 wins')
        
    elif p1.lower() == comp.lower():
        print('Tie!')
        
    else:
        print('computer picked', comp)
        print('computer wins')
        
rock_paper_sci()
```
